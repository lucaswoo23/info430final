```python
# Import libraries you would like to use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests 
from bs4 import BeautifulSoup as bs, SoupStrainer as ss
import re
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import os
chop = webdriver.ChromeOptions()
driver = webdriver.Chrome(chrome_options = chop)
# set webpage to all of rate my professors University of Washington professors
driver.get("https://www.ratemyprofessors.com/search.jsp?queryoption=TEACHER&queryBy=schoolDetails&schoolID=1530&schoolName=University+of+Washington&dept=select")
```

```python
# num loads is the number of professors divided by the number of professors shown on the page at a time. Need 201 loads to get to bottom
# 4030 professors / 20 per page = 202 - 1 because there's 20 professors to begin
numloads = 201
loadmore_counter = 0
counter = 0

y = 0 #y value will be used as a vertical position in the browser window
while loadmore_counter < numloads: #numloads is previously set based on how many reviews a school is listed to have    
    counter = counter + 1
    # adds a sleep counter because otherwise interception error occurs. 2 seconds in between scrolling and clicking load more
    time.sleep(2)
    #   loadmore = driver.find_element_by_xpath('//*[@data-school-id='+ sid + ']') #this is the xpath of the Load More Button
    loadmore = driver.find_element_by_css_selector('.progressbtnwrap .content')
    location  = loadmore.location_once_scrolled_into_view #dictionary with 'x' and 'y' keys associated with location values
    #the location_once_scrolled_into_view function returns how much further we need to scroll 
    y += location['y']  #so we add the y value returned to the previous y
    # Javascript code that goes down 
    driver.execute_script("document.getElementsByClassName('result-list').scrollDown += 500") #JavaScript command to scroll
    # clicks the loadmore
    loadmore.click()
    # increments load more counter
    loadmore_counter += 1
```

```python
# Get all links of professors after loading every professor
# create empty list
allLinks = []
# get all the links (a tags)
reviews = driver.find_elements_by_css_selector('.result-list a')
# for every link append the link to the allLinks list
for teacher in reviews:
#     print(teacher.get_attribute('href'))
    allLinks.append(teacher.get_attribute('href'))
# print(reviews)
```

```python
allLinks = allLinks[9:]
```

```python
allLinks = allLinks[1:]
allLinks
```

```python
# create dataframe for all professors
all_professors = pd.DataFrame(columns={'first_name', 'last_name', 'quality', 'would_take_again', 'difficulty'})
```

```python
for link in allLinks:
    page = requests.get(link)
    pageScrape = bs(page.content, 'html.parser')
    headline = pageScrape.find('div', class_ = 'headline')
    if (headline == None):
        all_professors = insertProfessor(pageScrape, all_professors)
    else:
        all_professors = insertEmptyProfessor(pageScrape, all_professors)
```

```python
# getting first name, last name, quality, difficulty, and would take again percentage
def insertProfessor(pageScrape, all_professors):
    firstName = pageScrape.find('span', class_ = 'pfname')
    firstName = firstName.text
    firstName = firstName.lstrip()
    firstName = firstName.rstrip()
    lastName = pageScrape.find('span', class_ = 'plname').text
    lastName = lastName.lstrip()
    lastName = lastName.rstrip()
    attributes = pageScrape.find_all('div', class_ = 'grade')
    quality = attributes[0].text
    would_take_again = attributes[1].text
    would_take_again = would_take_again.lstrip()
    would_take_again = would_take_again.rstrip()
    difficulty = attributes[2].text
    difficulty = difficulty.lstrip()
    difficulty = difficulty.rstrip()
    all_professors = all_professors.append({'first_name' : firstName , 'last_name' : lastName, 'quality' : quality, 'would_take_again' : would_take_again, 'difficulty' : difficulty} , ignore_index=True)
    return all_professors
```

```python
def insertEmptyProfessor(pageScrape, all_professors):
    professor = pageScrape.find('div', class_ = 'headline').text
    professor = professor[31:len(professor) - 1]
    professor = professor.split()
    firstName = professor[0]
    lastName = professor[1]
    all_professors = all_professors.append({'first_name' : firstName , 'last_name' : lastName, 'quality' : 'null', 'would_take_again' : 'null', 'difficulty' : 'null'} , ignore_index=True)   
    return all_professors
```

```python
all_professors.to_csv('rate_my_professors.csv', index = False)
```
